El codigo presentado a continuación analiza los datos del archivo BAO.dat y encuentra, usando la autocorrelacion de los datos, patrones repetitivos (que en este caso serian las medidas características de los círculos).

```python
%pylab inline
from scipy.fftpack import fft, fftfreq, ifft
#Carga los datos en un array
data= loadtxt('BAO.dat')
#Se define el array que contiene los datos y n como la cantidad de datos (29010)
data=array(data)
n=len(data[:,0])
scatter (data[0:n,0],data[0:n,1],alpha=0.2, s=0.6 )
print n
print 'Se puede observar en la grafica ciertos patrones circulares'
print 'Sin embargo, seria dificil estimar el radio dado el ruido presente, la distribucion irregular y el gráfico.'
print ''
```

Para estimar el radio de los círculos se usa la correlación de los datos. Dicho correlacion muestra si los datos presentan una tendencia en funcion de ellos mismos y cada cuanto lo hace. Sin embargo, para hacer el análisis en este caso es necesario generar una matriz que contenga como información el número de puntos presentes en cada cuadricula imaginaria ubicada en la gráfica de datos. AL obtener esta matriz, nos armamos con un elemento bidimensional que describe fielmente a los datos.

```python
#La matriz creada se llamará pixel. x_px y y_px definen las dimensiones de la matriz. En este caso de de (150, 150)
pixel=[]
count =-1
x_px, y_px=150,150
spa_x=max(data[:,0])-min(data[:,0])
spa_y=max(data[:,1])-min(data[:,1])
ini_x, ini_y=min(data[:,0]),max(data[:,1])
dx,dy=float(spa_x/x_px),float(spa_y/y_px)
for i in range(y_px):
    pixel.append([])
    count+=1
    for j in range(x_px):        
        num=size(data[:,0][(data[:,0]>ini_x+dx*i)&(data[:,0]<ini_x+dx*(i+1))&(data[:,1]<ini_y-dy*j)&(data[:,1]>ini_y-dy*(j+1))])
        pixel[count].append(num)
print shape(pixel)
imshow(pixel)
print 'En la imagen se aprecia un parecido razonable con los datos iniciales.'
print 'Esto nos permite confiar en la matriz creada.'
print ''
```

Los datos para calcular la autocorrelacion estan listos. Ahora, siguiendo el teorema de Wiene-Khinchin, la correlacion se calcula realizando la transformación de Fourier de los datos iniciales.

```python
fft_x = fft2(pixel,axes=(0,1))
```

Luego calculando el cuadrado absoluto de estra transformada multiplicando el resultado anterior por su complejo conjugado.

```python
inter= dot(conjugate(fft_x),fft_x)
```

Y luego realizando la antitransformada de Fourier de los datos resultantes.

```python
fft_x=ifft2(inter, axes=(0,1)).real
```

El siguiente es el resultado de dicho proceso.

```python
imshow(log(fft_x))
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
ax1 = fig.add_subplot(111,aspect='equal')
extent=(-3,3,-3,3)
im=ax1.imshow(log(fft_x), extent=extent)
```

El gráfico presenta cierto orden cuadriulado en toda su extensión. Aunque las dimensiones varian entre cuadrículas, se puede obserbar que la longitud dominante de aprox. 0.75 unidades. Esta longitud se presenta util para realizar análisis.

Si conocieramos exactamente la distribución de los circulos en los datos iniciales (todos juntados por sus extremos o cortados por sus mitades) podriamos relacionar exactamente las distancias de correlacion presentadas por la matriz al diametro o al radio de los círculos.

Si los círculos estuvieran juntados por sus extremos se puede afrimar que el radio de los círculos es la mitad de la distancia de correlacion dominante en el gráfico. Entonces r sería aprox. 0.33 unidades.

Por otra parte, si los círcuos estubienran cortados por sus mitades (o dicho de otra forma, el extremo de un círculo no se encuentra en el extremo del otro sino en su centro) se puede afirmar que el radio seria la distancia de correlación dominante. Entoces r seria aprox. 0.75 unidades.
