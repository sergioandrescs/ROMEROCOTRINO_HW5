### Importamos los datos

```python
data = pd.read_csv('sparse_sample_monthrg.dat', sep=' ', na_values=['-99'])
data = data.dropna()
data = np.array(data.T)
```

### Modificamos los datos

```python
#Creamos una nueva lista para tener valores de el total de manchas solares, y no el promedio de estas

nuevo = []
for i in range(1612,1996):
    index = where(data[0]==i)
    #print index
    if(len(index[0])!= 0):
        temp = data.T[index]
        #print temp
        nManchas = 0
        nDias = 0
        for j in range(len(temp)):
            nDias += temp[j][2]
            nManchas += temp[j][2]*temp[j][3]
        nuevo.append([i, nManchas/nDias])
        
#Mostramos los datos sin interpolar
        
nuevo = array(nuevo)
plot(nuevo.T[0],nuevo.T[1])
```

### Interpolación

```python
#Interpolamos los datos

hola = sc.interpolate.PchipInterpolator(nuevo.T[0], nuevo.T[1])
nuevo2 = []
for i in range(1612,1996):
    nuevo2.append([i,hola.__call__(i)])
nuevo2 = array(nuevo2)

#Mostramos los datos interpolados

plot(nuevo2.T[0], nuevo2.T[1])
```

### Transformada de Fourier

```python
#Manipulamos los datos para obtener el periodo solar

frec = fftfreq(len(nuevo2), 1)

T = 1/frec

nuevo3 = np.abs(fft(nuevo2.T[1]))
```
