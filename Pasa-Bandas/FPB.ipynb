```python
%pylab inline
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq

data, sampl_f, enc = audio.wavread("recording.wav")
data=array(data)
n=len(data)
dt=1/float(sampl_f)
plot (data)
print (data)
```

```python
time_array = arange(0, len(data)/float(sampl_f), dt)

plot(time_array[:], data[:])
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
fft_data = fft(data[:]) / n
freq = fftfreq(n, dt)
plt.plot(freq,abs(fft_data))
print freq
```

```python
f_cut_min=1000
f_cut_max=1900
fft_data[abs(freq)<f_cut_min]=0
fft_data[abs(freq)>f_cut_max]=0
plt.plot(freq,abs(fft_data))
xlim([-f_cut_max*2,f_cut_max*2])
```

```python
data=ifft(fft_data[:]*n)
plot(time_array[:], abs(data[:]))
xlabel("time(s)", fontsize=20)
ylabel("Amplitude", fontsize=20)
```

```python
# Create Format object with file format .wav, 24-bit encoding, and big endianness
fmt = audio.Format('wav', 'pcm16', 'big')
# Create file to be written to with a single channel and a sampling frequency of 11025Hz
f = audio.Sndfile('recording.wav', 'w', fmt, 1, sampl_f)
# Write the data to the file
f.write_frames(data[:])
f.close()
```
