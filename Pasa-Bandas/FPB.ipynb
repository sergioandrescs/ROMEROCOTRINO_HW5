4. Filtro Pasa Bandas

El siguiente código realiza el filtrado de ciertas bandas de frecuencia elegidas por el usuario sobre una grabacion. Para probar su funcionamiento se ha usado la grabación de dos notas producidas por un piano a tiempos diferentes y al mismo tiempo. Acompañando a este archivo se encuentra dicha grabación, junto con los resultados filtrados para cada frecuencia. Todos los formatos usados para generar los archivos de sonido serán .wav.

El codigo a presentar uso algunos aportes del código https://raw.github.com/forero/ComputationalMethods/master/notes/fourier.ipynb del surso de métodos computacionales.

```python
%pylab inline
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq
#Se lee el archivo de audio formando un array que lo describe. 
#Dicha lectura se hace por un muestre al archivo a 48000 Hz.
#sampl_f representa esta frecuencia de muestreo y enc muestra la modulacion del pulso. 
data, sampl_f, enc = audio.wavread("recording.wav")
data=array(data)
n=len(data)
dt=1/float(sampl_f)

print (sampl_f), enc
print ''
```

```python
#Para visualizar el archivo se crea un array de tiempo usando la frecuencia para obtener el periodo de separación
time_array = arange(0, (len(data)/float(sampl_f)), dt)

plot(time_array[0:n], data[0:n])
xlabel("$Tiempo(s)$", fontsize=20)
ylabel("$Amplitud$", fontsize=20)
```

```python
#Ahora se realiza un análisis de Fourier de los datos para identificar que frecuencias se encuentran presentes.
fft_data = fft(data[:]) / n
freq = fftfreq(n, dt)
plot(freq,abs(fft_data))
```

```python
#Dado que los resutados son simétricos, solo necesitamos trabajar con la mitad de los datos para encontrar
#frecuencias importantes
half_n = ceil(n/2.0)
fft_data_half = (2.0) * fft_data[:half_n]
freq_half = freq[:half_n]
plot(freq_half, abs(fft_data_half))
```

```python
#Ahora se encuentran los picos de frecuencia. En este caso se presentan 10 picos.
for i in range(10):
    index = argmax(fft_data_half)
    if( freq_half[index] != 0):
        print "Hay un maximo de frecuencia en: \n", (freq_half[index])
    fft_data_half[index] = 0
```

En este caso, para comprobar el correcto funcionamiento del filtro pasa bandasvamos a elimianr uno de los dos tonos presentes en el archivo de audio. EL resultado del segmento anterior muestra que exiten dos grupos de frecuencias mas importantes: una cercana a los 988 Hz y otra cercana a los 262 Hz.

Además de estas frecuencias se presentan otras como 130 Hz y 392 Hz. Dichas frecuencias complementan el sonido producido por el piano, pues hay que tener en cuenta de dicho instrumento no produce frecuencias puras sino una combinacion de una cuanta frecuencias juntas.

En el segmento siguiente crearemos un filtro que solo permita el paso de frecuencias entre 200 y 300 Hz. La diferencia entre los extremos del intervalo y el valor maximo de frcuencia permite pasar  frecuencias cercanas a la frecuencia máxima, lo que hace que el sonido generado presente caracteristicas similares (aunque no identicas) a la grabación inicial.

```python
f_cut_min=200
f_cut_max=300
fft_data[abs(freq)<f_cut_min]=0
fft_data[abs(freq)>f_cut_max]=0
plt.plot(freq,abs(fft_data))
xlim([-f_cut_max*2,f_cut_max*2])
print 'La grafica muestra frecuencias no nulas al interior de la banda usada.'
```

```python
#Para generar el sonido filtrado se realiza una antitransformada de Fourier 
#que muestra las caracteristicas de dicho sonido.
data=ifft(fft_data[:]*n)
plot(time_array[0:n], abs(data[0:n]))
xlabel("$Tiempo(s)$", fontsize=20)
ylabel("$Amplitud$", fontsize=20)
```

```python
#Finalmente el array de datos se convierte en un archivo de sonido usando audiolab
#En este caso, se crea un archivo llamado new2.wav.

fmt = audio.Format('wav', 'pcm16', 'big')
f = audio.Sndfile('new2.wav', 'w', fmt, 1, sampl_f)
f.write_frames(abs(data))
f.close()
```

Acompañando este código, se enuentran los archivos de sonido filtrados para cada tono presente en la grabación inicial. Aunque el sonido generado es diferente a la grabación inicial, la similitud es suficiente como para que un oido inexperto pueda apreciarla.
